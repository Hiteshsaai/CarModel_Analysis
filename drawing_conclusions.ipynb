### Q1: Are more unique models using alternative sources of fuel? By how much?
For the year 2008, the unique models uses Gasoline by 99.69% nearly almost all vechile used Gasoline
For the year 2018, the unique models uses Gasoline by 75.88% and compared to 2008 it has reduced in percentage

```python
gasoline = len(df_08.query('fuel == "Gasoline"'))
cng = len(df_08.query('fuel == "cng"'))
ethonal = len(df_08.query('fuel == "ethonal"'))
gas = len(df_08.query('fuel == "gas"'))
gaso_per = (gasoline/987) * 100
cng_per = (cng/987) * 100
ethonal_per = (ethonal/987) * 100
gas_per = (gas/987) * 100
print ( gaso_per,cng_per,ethonal_per,gas_per)
count=[1,2,3,4]
hights=[gaso_per,cng_per,ethonal_per,gas_per]
labels=['gasoline','cng','ethonal','gas']
plt.bar(count,hights,tick_label= labels)
plt.title("Percentage use of Fuel's By Vechile")
plt.xlabel("Fuels")
plt.ylabel("Percentage")
plt.show()
```

```python
gasoline1 = len(df_18.query('fuel == "Gasoline"'))
cng1 = len(df_18.query('fuel == "cng"'))
ethonal1 = len(df_18.query('fuel == "ethonal"'))
gas1 = len(df_18.query('fuel == "gas"'))
gaso_per1 = (gasoline1/987) * 100
cng_per1 = (cng1/987) * 100
ethonal_per1 = (ethonal1/987) * 100
gas_per1 = (gas1/987) * 100
print ( gaso_per1,cng_per1,ethonal_per1,gas_per1)
count=[1,2,3,4]
hights=[gaso_per1,cng_per1,ethonal_per1,gas_per1]
labels=['gasoline','cng','ethonal','gas']
plt.bar(count,hights,tick_label= labels)
plt.title("Percentage use of Fuel's By Vechile")
plt.xlabel("Fuels")
plt.ylabel("Percentage")
plt.show()
```

### Q2: How much have vehicle classes improved in fuel economy?  
Except for small car and station wagon all other vehicle class have improved the fuel econom

```python
veh_08 = df_08.groupby('veh_class',as_index= False).mean()
plt.subplots(figsize=(10,6))
sns.barplot(x='veh_class',y='cmb_mpg',data = veh_08)
plt.show()
veh_18 = df_18.groupby('veh_class',as_index= False).mean()
plt.subplots(figsize=(10,6))
sns.barplot(x='veh_class',y='cmb_mpg',data = veh_18)
plt.show()
```
